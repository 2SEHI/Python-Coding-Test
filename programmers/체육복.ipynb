```python
def solution(n, lost, reserve):
    # 여분의 체육복이 없지만 도난받은 학생
    lost_reverse = [i for i in lost if i not in reserve]
    # print(lost_reverse)
    # 여분의 체육복이 있고 도난당하지 않은 학생
    reverse_lost = [i for i in reserve if i not in lost]
    # print(reverse_lost)

    answer = 0
    if len(lost_reverse) - len(reverse_lost) < 0 :
        answer = n
    else :
        answer = n - len(lost_reverse) + len(reverse_lost)
    return answer

print(solution(2, [1, 2], [1, 2]))
print(solution(2, [1, 2], [1]))
print(solution(2, [1, 2], [2]))
print(solution(2, [1], [1, 2]))
print(solution(2, [1], [1]))
print(solution(2, [1], [2]))
print(solution(2, [2], [1, 2]))
print(solution(2, [2], [1]))
print(solution(2, [2], [2]))
```
